# 회귀 분석이란 무엇인가요?

회귀 분석은 종속 변수('결과 변수' 또는 '대상'이라고도 함)와 하나 이상의 독립 변수('특징', '예측자' 또는 '공변량'이라고도 함) 간의 관계를 
추정하기 위한 일련의 통계 프로세스입니다.
머신 러닝에서 회귀의 목적은 판매 금액, 수량, 온도 등과 같은 연속적인 값을 예측하는 것입니다.

## 회귀분석 알고리즘 모듈 사용하기 

```python
from pycaret.regression import *
```

## setup 매개변수

- session_id: 추후 재현성을 위해 모든 함수에 시드로 배포되는 시드 난수입니다. 세션_id를 전달하지 않으면 자동으로 난수가 생성되어 모든 함수에 배포됩니다. 이 실험에서는 추후 재현성을 위해 세션_id를 123으로 설정했습니다.

- 원본 데이터 : 데이터 세트의 원래 모양을 표시합니다. 이 실험에서 (5400, 8)은 대상 열을 포함한 5400개의 샘플과 8개의 특징을 의미합니다.

누락된 값 : 원본 데이터에 누락된 값이 있는 경우 True로 표시됩니다. 이 실험에서는 데이터 집합에 결측값이 없습니다.

- 숫자 특징 : 숫자로 추론된 특징의 수입니다. 이 데이터 세트에서는 8개 특징 중 1개가 숫자로 추론됩니다.

- 범주형 특징 : 범주로 추론된 특징의 개수입니다. 이 데이터셋에서는 8개 특징 중 6개가 범주형으로 추론됩니다.

- 변환된 훈련 세트 : 변환된 훈련 세트의 모양을 표시합니다. 변형된 훈련 집합의 경우 (5400, 8)의 원래 모양이 (3779, 28)로 변형된 것을 확인할 수 있습니다. 범주형 인코딩으로 인해 특징의 수가 28개에서 8개로 증가했습니다.

- 변환된 테스트 세트: 변환된 테스트/홀드아웃 세트의 모양을 표시합니다. 테스트/홀드아웃 세트에는 1621개의 샘플이 있습니다. 이 분할은 기본값인 70/30을 기준으로 하며, 설정에서 train_size 파라미터를 사용하여 변경할 수 있습니다.
