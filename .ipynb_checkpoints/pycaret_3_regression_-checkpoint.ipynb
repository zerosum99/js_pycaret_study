```python
### load sample dataset from pycaret dataset module
from pycaret.datasets import get_data
data = get_data('insurance')
```

```python
data.shape
```

## Setup


설정 함수는 훈련 환경을 초기화하고 변환 파이프라인을 생성합니다. 
PyCaret에서 다른 함수를 실행하기 전에 설정 함수를 호출해야 합니다. 
이 함수에는 데이터와 대상이라는 두 가지 필수 매개변수만 있습니다. 다른 모든 파라미터는 선택 사항입니다.

```python
# import pycaret regression and init setup
from pycaret.regression import *
s = setup(data, target = 'charges', session_id = 123)
```

## PyCaret에는 작업할 수 있는 두 가지 API 세트가 있습니다. 

- (1) 함수형(위에서 본 것처럼) 및 (2) 객체 지향 API.

함수를 직접 실행하는 대신 객체 지향 API를 사용하면 클래스를 가져와서 클래스의 메서드를 실행합니다.

```python
# import RegressionExperiment and init the class
from pycaret.regression import RegressionExperiment
exp = RegressionExperiment()
```

```python
exp
```

```python
# init setup on exp
exp.setup(data, target = 'charges', session_id = 123)
```

## Compare Models

- compare_models 함수는 교차 검증을 사용하여 모델 라이브러리에서 사용할 수 있는 모든 추정기의 성능을 학습하고 평가합니다. 

- 이 함수의 출력은 교차 검증 평균 점수가 포함된 점수 그리드입니다. CV 중에 평가된 지표는 get_metrics 함수를 사용하여 액세스할 수 있습니다. 

- 사용자 지정 지표는 add_metric 및 remove_metric 함수를 사용하여 추가하거나 제거할 수 있습니다.


```python
# compare baseline models
best = compare_models()
```

```python
best
```

```python
# compare models using OOP
exp.compare_models()
```

## Analyze Model

- plot_model 함수는 테스트 세트에서 학습된 모델의 성능을 분석하는 데 사용됩니다. 

- 특정 경우에는 모델을 다시 훈련해야 할 수도 있습니다.

```python
# plot residuals
plot_model(best, plot = 'residuals')
```

```python
# plot error
plot_model(best, plot = 'error')
```

```python
# plot feature importance
plot_model(best, plot = 'feature')
```

```python
# check docstring to see available plots 
# help(plot_model)
```

```python
evaluate_model(best)
```

## Prediction

- 예측 모델 함수는 예측_라벨을 입력 데이터 프레임에 새 열로 반환합니다. 

- 데이터가 없음(기본값)인 경우, 설정 함수 중에 생성된 테스트 집합을 사용하여 점수를 매깁니다.
